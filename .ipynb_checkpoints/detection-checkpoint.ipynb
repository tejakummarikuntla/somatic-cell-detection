```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
image = cv2.imread('somatic_deep.jpg')
image_copy = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_copy)
```

```python
# RGB Channels

r = image_copy[:, :, 0]
g = image_copy[:, :, 1]
b = image_copy[:, :, 2]

f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20,10))

ax1.set_title('Red')
ax1.imshow(r, cmap='gray')

ax2.set_title('Green')
ax2.imshow(g, cmap='gray')

ax3.set_title('Blue')
ax3.imshow(b, cmap='gray')
```

```python
# Converiting from RGB to HSV 

hsv = cv2.cvtColor(image_copy, cv2.COLOR_RGB2HSV)
plt.imshow(hsv)
```

```python
h = hsv[:, :, 0]
s = hsv[:, :, 1]
v = hsv[:, :, 2]

f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20,10))

ax1.set_title('Hue')
ax1.imshow(h, cmap='gray')

ax2.set_title('Saturation')
ax2.imshow(s, cmap='gray')

ax3.set_title('Value')
ax3.imshow(v, cmap='gray')
```

```python
# Define Color selection in RGB values

low_blue = np.array([10, 90, 183])
upper_blue = np.array([225, 225, 230])
```

```python
# Define Color selection in HSV values

low_hue = np.array([160, 0, 0])
upper_hue = np.array([180, 255, 255])
```

```python
# RGB Mask

mask_rgb = cv2.inRange(image_copy, low_blue, upper_blue)

masked_image = np.copy(image_copy)
masked_image[mask_rgb == 0] = [0, 0, 0]

plt.imshow(masked_image)
plt.title('RGB Selection')
```

```python
mask_hsv = cv2.inRange(hsv, low_hue, upper_hue)

masked_image = np.copy(image_copy)
masked_image[mask_hsv == 0] = [0, 0, 0]

plt.imshow(masked_image)
plt.title('HSV Selection')
```
